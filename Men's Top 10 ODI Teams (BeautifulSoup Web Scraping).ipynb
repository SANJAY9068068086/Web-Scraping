# Top 10 ODI Man's Team (BeautifulSoup Web Scraping)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
page = requests.get(url)
soup = BeautifulSoup(page.content)
soup.prettify()
```

```python
team_name = [] 
total_match = [] 
total_points = [] 
ratings = []  
```

```python
team = soup.find_all('span',class_="u-hide-phablet")
for i in team:
    team_name.append(i.text)
name = team_name[:10]
```

```python
match = soup.find('td',class_="rankings-block__banner--matches")
match.text
total_match.append(match.text)
```

```python
info = []
match_1 = soup.find_all('td',class_="table-body__cell u-center-text")
for i in match_1:
    info.append(i.text)
for i in range(0,len(info),2):
    total_match.append(info[i])
mat = total_match[:10]
```

```python
points = soup.find('td',class_="rankings-block__banner--points")
points.text
total_points.append(points.text)
```

```python
for i in range(1,len(info),2):
    total_points.append(info[i])
poi = total_points[:10]
```

```python
rating = soup.find('td',class_="rankings-block__banner--rating u-text-right")
ratings.append(rating.text.replace("\n","").strip())
```

```python
info_1 = []
rating_1 = soup.find_all('td',class_="table-body__cell u-text-right rating")
for i in rating_1:
    info_1.append(i.text.replace("\n",""))
for j in range(0,len(info_1)):
    ratings.append(info_1[j])
rat = ratings[:10]
```

```python
Top_10_ODI_Mans = pd.DataFrame(index=list(range(1,11)))
Top_10_ODI_Mans["Team Name"] = name
Top_10_ODI_Mans["Total Matches"] = mat
Top_10_ODI_Mans["Total Points"] = poi
Top_10_ODI_Mans["Ratings"] = rat
```
