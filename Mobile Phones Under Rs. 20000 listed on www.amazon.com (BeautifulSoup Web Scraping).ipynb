## Mobile Phones Under Rs. 20000 on www.amazon.com (BeautifulSoup Web Scraping)

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
page = requests.get("https://www.amazon.in/s?k=mobiles+phones+under+20000&i=electronics&rh=n%3A1389432031%2Cp_36%3A300000-2000000&dc&qid=1627255011&rnid=1318502031&ref=sr_nr_p_36_1%22")
soup = BeautifulSoup(page.content)
page
```

```python
product_name = []
price= []
image_url = []
average_rating = []
```

```python
product = soup.find_all("span",class_="a-size-medium a-color-base a-text-normal")
for i in product:
    product_name.append(i.text)
```

```python
pri = soup.find_all("span",class_="a-price-whole")
for i in pri:
    price.append(i.text)
```

```python
rating = soup.find_all("span",class_="a-icon-alt")
for i in rating:
    average_rating.append(i.text)
avg_rating = average_rating[:24]
```

```python
image = soup.find_all("img",class_="s-image")
for i in image:
    image_url.append(i.get("src"))
```

## Mobile Phones Under Rs. 20000 

```python
Mobile_Phone_Description = pd.DataFrame(index = list(range(1,25)))
Mobile_Phone_Description["Product Name"] = product_name
Mobile_Phone_Description["Price"] = price
Mobile_Phone_Description["Average Rating"] = avg_rating
Mobile_Phone_Description["Image URL"] = image_url

Mobile_Phone_Description
```
